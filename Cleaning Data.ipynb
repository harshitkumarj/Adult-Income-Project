# Working on Occupation

## Filling the workclass

### w is test

```python
w=df[df['occupation'].isna()]
```

```python
w.shape
```

```python
w=w.drop(axis=1,labels=['workclass'])
```

```python
w.head()
```

```python
features=w.select_dtypes(include=object).columns.tolist()
for i in features:
    l=w[i].unique().tolist()
    j=[int(p) for p in range(0,len(l))]
    w[i].replace(to_replace=l,value=j,inplace=True)
```

```python
w=w.drop(['fnlwgt','marital-status','race','native-country','income'], axis=1)
```

```python
w_dep = w['occupation']
```

```python
w_ind = w.drop(axis=1,labels=['occupation'])
```

```python
w_ind.head()
```

### v is train

```python
v=df.dropna()
```

```python
v.shape
```

```python
features=v.select_dtypes(include=object).columns.tolist()
for i in features:
    l=v[i].unique().tolist()
    j=[int(p) for p in range(0,len(l))]
    v[i].replace(to_replace=l,value=j,inplace=True)
```

```python
v=v.drop(['workclass','fnlwgt','marital-status','race','native-country','income'], axis=1)
```

```python
v.head()
```

```python
v_dep = v['occupation']
```

```python
v_ind = v.drop(axis=1,labels=['occupation'])
```

```python
v_ind.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
# evaluate each model in turn
results = []
names = []
for name, model in models:
    kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
    cv_results = cross_val_score(model, v_ind, v_dep, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

fill native-country on the basis of race

```python
model = LogisticRegression(solver='liblinear', multi_class='ovr')
model.fit(v_ind, v_dep)
```

```python
prediction = model.predict(w_ind)
```

```python
prediction
```

```python
w_ind.head()
```

```python
w_ind['occupation']=prediction
```

```python
w_ind.head()
```

```python
w_ind.info()
```

## Replacing the nan values with obtained values

```python
data=df
data.head()
```

```python
features=data.select_dtypes(include=object).columns.tolist()
for i in features:
    l=data[i].unique().tolist()
    j=[int(p) for p in range(0,len(l))]
    data[i].replace(to_replace=l,value=j,inplace=True)
```

```python
data.head()
```

replacing 3 with obtained values because in occupation column, category 3 is NaN

```python
for i in range(len(data)):
    if data['occupation'][i]==3:
        data['occupation'][i]=w_ind['occupation'][i]
```

```python
data.head()
```
